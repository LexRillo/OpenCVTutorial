# Averaging blur

```python
%matplotlib inline
import os 
import numpy as np
import cv2
from matplotlib import pyplot as plt

pic = cv2.imread('meerkat.jpg')

kernel = np.ones((5,5),np.float32)/25
blurred = cv2.filter2D(pic,-1,kernel)

plt.imshow(blurred)
plt.show()
```

# Gaussian blur

```python
pic = cv2.imread('meerkat.jpg')
matrix = (7,7)

blurred = cv2.GaussianBlur(pic,matrix, 0)

plt.imshow(blurred)
plt.show()
```

# Median blur

```python
pic = cv2.imread('meerkat.jpg')

blurred = cv2.medianBlur(pic,3)

plt.imshow(blurred)
plt.show()
```

# Bilateral filter

```python
pic = cv2.imread('meerkat.jpg')

dimpixel = 7
color = 100
space = 100

filtered = cv2.bilateralFilter(pic,dimpixel,color,space)

plt.imshow(filtered)
plt.show()
```

# Canny edge detection

```python
pic = cv2.imread('meerkat.jpg')

thresholdval1 = 100
thresholdval2 = 200

canny = cv2.Canny(pic, thresholdval1, thresholdval2)

plt.imshow(canny)
plt.show()
```
